## ***4. Projections and Predictions:***

○ Based on the generated data, make a data-driven projection about lead conversion rates and suggest a budget allocation strategy for marketing and targeting.

1. Data Preparation: (Optional) Add a 'Converted' column:

* You'll need to either simulate conversion data based on assumptions or use
actual historical data if available.

* For this example, let's simulate conversion data by randomly assigning a 'Converted' status (True/False) to each lead.

```python
conversion_rates = [0.05, 0.1, 0.02, 0.08, 0.03]  # Example conversion rates for different lead sources
conversion_probabilities = np.random.choice(conversion_rates, size=len(df))
df['Converted'] = np.random.rand(len(df)) < conversion_probabilities
```

2. Analyze Conversion Rates:

* Overall Conversion Rate:

```python
overall_conversion_rate = df['Converted'].mean()
print(f"Overall Conversion Rate: {overall_conversion_rate:.2%}")
```

* Conversion Rates by Lead Source:

```python
conversion_rates_by_source = df.groupby('Lead Source')['Converted'].mean()
print(conversion_rates_by_source)
```

* Visualize Conversion Rates:

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=conversion_rates_by_source.index, y=conversion_rates_by_source)
plt.title("Conversion Rates by Lead Source")
plt.xlabel("Lead Source")
plt.ylabel("Conversion Rate")
plt.xticks(rotation=45)
plt.show()
```

3. Budget Allocation Strategy:
* Calculate Cost per Lead (CPL) for each source:

```python
cpl_per_source = {
    'Instagram': 0.5,  # Example CPL values
    'LinkedIn': 1.0,
    'College Collaboration': 0.2,
    'Google Form': 0.1,
    'Mass-Mailing': 0.05,
    'Whatsapp': 0.01
}
```

* Calculate Cost per Conversion (CPC) for each source:

```python
cpc_per_source = {}
for source, conversion_rate in conversion_rates_by_source.items():
    cpc_per_source[source] = cpl_per_source[source] / conversion_rate
```

* Allocate Budget Proportionally:

```python
total_budget = 10000  # Example budget

budget_allocation = {}
for source, cpc in cpc_per_source.items():
    budget_allocation[source] = (1 / cpc) / sum(1 / cpc for cpc in cpc_per_source.values()) * total_budget

print(budget_allocation)
```

4. Projections and Predictions:

* Project Future Conversions:

```python
projected_conversions = {}
for source, budget in budget_allocation.items():
    projected_leads = budget / cpl_per_source[source]
    projected_conversions[source] = projected_leads * conversion_rates_by_source[source]

print(projected_conversions)
```
